**1) Importación de librería utilizadas en las funciones creadas**

```python
import requests
from bs4 import BeautifulSoup
import json
import csv
import gc
import re
import time
```

**2) Función utilizada para comprobar la existencia de una palabra dentro de una cadena**

```python
def get_data_into_section_document(word, docxf):
    """
        Parámetros de entrada
            word  String: Palabra buscada
            docxf String: Cadena de texto en donde se va a buscar
            
        Parámetro de salida
            _valueout: Cadena con el valor de la palabra buscada
    """
    
    _valueout=""
    
    if (docxf.find(word) != -1):
        _valueout=docxf.replace(word, '')
    else:
        _valueout=""
        
    return _valueout
```

**3) Función utilizada para obtener los datos desde la página web de Dialnet y guardarlos en un archivo csv**

```python
def get_data_of_dialnet(keyword, limit, numpapers, csv_writer):
    """
        Parámetros de entrada:
            keyword    String: Término de búsqueda
            limit      int:    Número de documentos por búsqueda
            numpapers  int:    Número total de artículos que se desea obtener
            csv_writer Object: Objeto relacionado al archivo csv
    """
    
    _valoffset=0
    _vallimit=limit
    _valnumarticulos=numpapers
    _elements = ['ISSN-e', 'ISSN', 'ISBN', 'Vol.', 'Nº.', 'págs.']
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/38.0.2125.122 UBrowser/4.0.3214.0 Safari/537.36'}
    
    try:
        
        while _valnumarticulos>0:
            _urlwbsc="https://dialnet.unirioja.es/buscar/documentos?querysDismax.DOCUMENTAL_TODO="+str(keyword)+"&registrosPorPagina="+str(_vallimit)+"&camposOrdenacion=%7BDOCUMENTAL_SORT_SCORE%3DDESC%7D&inicio="+str(_valoffset)
            _webpagesc=requests.get(_urlwbsc, headers=headers)
            _result_obtn = BeautifulSoup(_webpagesc.content)
            
            _artic_imp = _result_obtn.find_all('li', class_="articulo impar")
            _artic_par = _result_obtn.find_all('li', class_="articulo par")
            
            _contador=0
            while _contador<len(_artic_imp):
                _contloc=0
                _issne=""
                _issn=""
                _isbn=""
                _vol=""
                _numb=""
                _pags=""
                _autores=""
                
                if (_artic_imp[_contador].find(class_="descripcion").find(class_="autores")) is not None:
                    _autores=str(_artic_imp[_contador].find(class_="descripcion").find(class_="autores").get_text())
                    
                if (_artic_imp[_contador].find(class_="descripcion").find(class_="localizacion")) is not None:
                    _datos_loc=_artic_imp[_contador].find(class_="descripcion").find(class_="localizacion").get_text().split(',')
                    
                    
                    for _elemt in _elements:
                        _contloc=0
                        while _contloc<len(_datos_loc):
                            if _elemt=="ISSN-e":
                                _issne=get_data_into_section_document("ISSN-e",_datos_loc[_contloc]).strip();
                            elif _elemt=="ISSN":
                                _issn=get_data_into_section_document("ISSN",_datos_loc[_contloc]).strip();
                            elif _elemt=="ISBN":
                                _isbn=get_data_into_section_document("ISBN",_datos_loc[_contloc]).strip();
                            elif _elemt=="Vol.":
                                _vol=get_data_into_section_document("Vol.",_datos_loc[_contloc]).strip();
                            elif _elemt=="Nº.":
                                _numb=get_data_into_section_document("Nº.",_datos_loc[_contloc]).strip();
                            else:
                                _pags=get_data_into_section_document("págs.",_datos_loc[_contloc]).strip();

                            _contloc+=1
                
                
                _datapp=["Artículo",str(_artic_imp[_contador].find(class_="descripcion").find(class_="titulo").get_text()),
                         _autores,_datos_loc[0],_issne,_issn,_isbn,_vol,_numb,_pags]
                csv_writer.writerow(_datapp)
                _contador+=1
                
            
            _contador=0
            while _contador<len(_artic_par):
                _contloc=0
                _issne=""
                _issn=""
                _isbn=""
                _vol=""
                _numb=""
                _pags=""
                _autores=""
                
                if (_artic_par[_contador].find(class_="descripcion").find(class_="autores")) is not None:
                    _autores=str(_artic_par[_contador].find(class_="descripcion").find(class_="autores").get_text())
                    
                if (_artic_par[_contador].find(class_="descripcion").find(class_="localizacion")) is not None:
                    _datos_loc=_artic_par[_contador].find(class_="descripcion").find(class_="localizacion").get_text().split(',')
                    
                    for _elemt in _elements:
                        _contloc=0
                        while _contloc<len(_datos_loc):
                            if _elemt=="ISSN-e":
                                _issne=get_data_into_section_document("ISSN-e",_datos_loc[_contloc]).strip();
                            elif _elemt=="ISSN":
                                _issn=get_data_into_section_document("ISSN",_datos_loc[_contloc]).strip();
                            elif _elemt=="ISBN":
                                _isbn=get_data_into_section_document("ISBN",_datos_loc[_contloc]).strip();
                            elif _elemt=="Vol.":
                                _vol=get_data_into_section_document("Vol.",_datos_loc[_contloc]).strip();
                            elif _elemt=="Nº.":
                                _numb=get_data_into_section_document("Nº.",_datos_loc[_contloc]).strip();
                            else:
                                _pags=get_data_into_section_document("págs.",_datos_loc[_contloc]).strip();

                            _contloc+=1
                
                
                _datapp=["Artículo",str(_artic_par[_contador].find(class_="descripcion").find(class_="titulo").get_text()),
                         _autores,_datos_loc[0],_issne,_issn,_isbn,_vol,_numb,_pags]
                csv_writer.writerow(_datapp)
                _contador+=1
                
            
            _valoffset+=(_vallimit+1)
            _valnumarticulos-=_vallimit

    except requests.exceptions.HTTPError as errh:
        print(errh)
    except requests.exceptions.ConnectionError as errc:
        print(errc)
    except requests.exceptions.Timeout as errt:
        print(errt)
    except requests.exceptions.RequestException as err:
        print(err) 
```

**4) Función principal de la búsqueda, captura y recopilación de los datos bibliográficos de documentos científicos**

```python
def get_data_of_documents(limits, numpapers):
    """
        Parámetros de entrada:
            limits     int: Número de documentos por búsqueda
            numpapers  int: Número total de artículos que se desea obtener
    """
    
    _listpapers=[]
    _datahead = ['Fuente', 'Título', 'Autor', 'Editorial', 'ISSN-e', 'ISSN', 'ISBN', 'Volumén','Número','Páginas']
    _listkeywords=['data+science','data+mining','web+mining','web+scraping','text+mining','data+visualization']
    
    csv_file =  open("_dtpapersdnt.csv", 'w', newline='', encoding="utf-8")
    csv_writer = csv.writer(csv_file, delimiter=",", quoting=csv.QUOTE_ALL)
    csv_writer.writerow(_datahead)
    
    for _keyword in _listkeywords:
        get_data_of_dialnet(_keyword, limits, numpapers, csv_writer)
        time.sleep(5)

    csv_file.close()
```
